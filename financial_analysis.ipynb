```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas as pd
import seaborn as sns
color = sns.color_palette()
import sklearn.metrics as metrics

import warnings
warnings.filterwarnings('ignore')
# Load the dataset
file_path = '/content/Financial Analytics data (2).csv'
data = pd.read_csv(file_path)
```

```python
# Remove the unnamed column
data_cleaned = data.drop(columns=['Unnamed: 4'])
```

```python
# Display basic statistics of the cleaned dataset
data_cleaned.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot to visualize the relationship between Market Capitalization and Quarterly Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=data_cleaned)
plt.title('Market Capitalization vs Quarterly Sales')
plt.xlabel('Market Capitalization (Crores)')
plt.ylabel('Quarterly Sales (Crores)')
plt.show()

# Calculate the correlation between Market Capitalization and Quarterly Sales
correlation = data_cleaned[['Mar Cap - Crore', 'Sales Qtr - Crore']].corr()
print(correlation)
```

```python
# Top 10 Companies by Market Capitalization
top_10_market_cap = data_cleaned.nlargest(10, 'Mar Cap - Crore')

# Top 10 Companies by Quarterly Sales
top_10_quarterly_sales = data_cleaned.nlargest(10, 'Sales Qtr - Crore')
```

```python
# Distribution of Market Capitalization
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Mar Cap - Crore'], bins=30, kde=True)
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (Crores)')
plt.ylabel('Frequency')
plt.show()

# Distribution of Quarterly Sales
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Sales Qtr - Crore'], bins=30, kde=True)
plt.title('Distribution of Quarterly Sales')
plt.xlabel('Quarterly Sales (Crores)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for Market Capitalization
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_cleaned['Mar Cap - Crore'])
plt.title('Box Plot of Market Capitalization')
plt.xlabel('Market Capitalization (Crores)')
plt.show()

# Box plot for Quarterly Sales
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_cleaned['Sales Qtr - Crore'])
plt.title('Box Plot of Quarterly Sales')
plt.xlabel('Quarterly Sales (Crores)')
plt.show()
```

```python
# Pair plot to visualize relationships between variables
sns.pairplot(data_cleaned[['Mar Cap - Crore', 'Sales Qtr - Crore']])
plt.show()
```
